```python
# ---
# title: "Practico: Tablas de Crecimiento"
# author: "Carranza"
# date: "5/26/2018"
# ---

# %matplotlib notebook
%matplotlib inline
import pandas as pd
import numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Grow table - Weight for Age
weight_age = pd.read_csv('../../datasets/wtageinf.csv')
weight_age[:100:10]
```

```python
# Type checking
weight_age.dtypes
```

Se puede ver que los tipos de datos para las columnas de interes son números flotantes por lo cual es posible continuar con la construcción de las tablas de percentiles.

```python
# Useful constants and functions
percentiles = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']

def get_weight_by_age_percentiles_plot(data):
    for p in percentiles:
        plt.plot(data['Agemos'], data[[p]], label=p)

    plt.axis(xmin=0, xmax=36)
    plt.xlabel('Mes')
    plt.ylabel('Peso (Kg)')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
_, ax_freq = plt.subplots(figsize=(10, 10))
ax_freq.set_title('Percentiles de peso por edad (niños)')

# Boys Weight per age percentile chart
boys_weight_data = weight_age[weight_age.Sex == 1]
get_weight_by_age_percentiles_plot(boys_weight_data)
```

```python
_, ax_freq = plt.subplots(figsize=(10, 10))
ax_freq.set_title('Percentiles de peso por edad (niñas)')

# Girls Weight per age percentile chart
girls_weight_data = weight_age[weight_age.Sex == 2]
get_weight_by_age_percentiles_plot(girls_weight_data)
```

```python
# Grow table - Length for Age
length_age = pd.read_csv('../../datasets/lenageinf.csv')
length_age[:100:10]
```

```python
# Let's check the type for each column
length_age.dtypes
```

Podemos observar de la tabla anterior que existe un valor para el sexo que no corresponde con los valores 1 y 2.

```python
length_age.Sex.unique()
```

```python
length_age[length_age.Sex == "Sex"]
```

En efecto podemos observar que hay una fila que introduce ruido en los datos originales, procederemos a eliminarla

```python
# Discard corrupted row
filtered_length_age = length_age.drop(length_age.index[37])
filtered_length_age.Sex.unique()
```

Otra observación es que los valores de la columna "Agemos" como así también los valores que representan a los percentiles son de tipo `object`. Por lo cual, se procederá a transformar dichas columnas en números decimales para posteriormente poder trabajar con los mismos.

```python
percentiles = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']
converted_length_age = filtered_length_age[['Sex', 'Agemos'] + percentiles].convert_objects(convert_numeric=True)
converted_length_age
```

```python
# Useful constants and functions
percentiles = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']

def get_length_by_age_percentiles_plot(data):
    for p in percentiles:
        plt.plot(data['Agemos'], data[[p]], label=p)

    plt.axis(xmin=0, xmax=36)
    plt.xlabel('Mes')
    plt.ylabel('Longitud (cm)')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
_, ax_freq = plt.subplots(figsize=(10, 10))
ax_freq.set_title('Percentiles de longitud por edad (niños)')

# Boys Length per age percentile chart
boys_length_data = converted_length_age[converted_length_age.Sex == 1]
get_length_by_age_percentiles_plot(boys_length_data)
```

```python
_, ax_freq = plt.subplots(figsize=(10, 10))
ax_freq.set_title('Percentiles de longitud por edad (niñas)')

# Girls Length per age percentile chart
girls_length_data = converted_length_age[converted_length_age.Sex == 2]
get_length_by_age_percentiles_plot(girls_length_data)
```
